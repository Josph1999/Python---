```python
#გვიჩვენებს მონეტებს

coins = [1,5,10,20,50]

#აქ იწერება თანხის რაოდენობა

n = input("გთხოვთ შეიყვანოთ თანხა: ")

#გადაჰყავს მონაცემი string-დან integer-ში

n_int=int(n)

#ფუნქცია რომელიც აჩვენებს მონეტებსა და თანხას

def min_coins(coins,n_int):
    table = [0] + [float('inf')]*n_int
    
    #გვიჩვენებს უსასრულო დიაპაზონს
    
    for coin in coins:
        for i in range(coin,n_int+1):
            table[i] = min(table[i],table[i-coin]+1)
            
            #გადააქვს საბოლოო შედეგი
            
    return table[n_int] 

#გვიჩვენებს საბოლოო შედეგს

print(min_coins(coins,n_int)); print("არის მინიმალური მონეტ"("ებ")"ის რაოდენობა")

#თავიდან იწყებს ფუნქციას()

while True:
    n = input("გთხოვთ შეიყვანოთ თანხა: ")

    n_int=int(n)

    def min_coins(coins,n_int):
        table = [0] + [float('inf')]*n_int
    
        for coin in coins:
            for i in range(coin,n_int+1):
                table[i] = min(table[i],table[i-coin]+1)
            
            
        return table[n_int] 

    print(min_coins(coins,n_int)); print("არის მინიმალური მონეტ"("ებ")"ის რაოდენობა")
    

    
```
